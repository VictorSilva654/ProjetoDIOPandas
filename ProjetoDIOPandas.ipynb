```python
#Importando as librarys
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("grayscale")
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Total de colunas e linhas
df.shape
```

```python
#Renomeando as colunas
df = df.rename(columns={"SalesOrderLineKey": "PedidoVendas", "ResellerKey":"Revendedor", "CustomerKey":"Consumidor", "ProductKey":"Produto", "OrderDateKey":"DataPedido","DueDateKey":"DataVencimento", "ShipDateKey":"DataEntrega", "SalesTerritoryKey":"VendasRegião", "Order Quantity":"Quantidade","Unit Price":"PreçoUnidade", "Extended Amount":"TotalUnidades", "Unit Price Discount Pct":"DescontoPorUnidade", "Product Standard Cost":"CustoPorUnidade","Total Product Cost": "PreçoTotalDeCusto", "Sales Amount":"ValorTotalVenda"})
```

```python
#Verificando tipos de dados
df.dtypes
```

```python
df.describe()
```

```python
df["Revendedor"].unique()
```

```python
#Recuperando vendas de um vendedor pelo loc
vendedorProduto = df.loc[df["Revendedor"] == 408]
vendedorProduto.head()
```

```python
#Trazendo a média do preço de custo dos produtos
df["PreçoTotalDeCusto"].mean()
```

```python
#Vendo valores nulos na tabela
df.isnull().sum()
```

```python
df.sample(14)
```

```python
#Retirando valores nulos da tabela
df.dropna(subset=["DataEntrega"],inplace=True)
```

```python
df.dtypes
```

```python
#Transformando os dados em data
df["DataPedido"] = pd.to_datetime(df["DataPedido"], format='%Y%m%d')
df["DataVencimento"] = pd.to_datetime(df["DataVencimento"], format='%Y%m%d')
df["DataEntrega"] = pd.to_datetime(df["DataEntrega"], format='%Y%m%d')
```

```python
#Tiragem de 9 registros
df.sample(9)
```

```python
#Diferença entre a data de entrega e a data de pedido
df["DiferencaDiasPedido"] = df["DataEntrega"] - df["DataPedido"] 
```

```python
vendas_janeiro_2020 = df.loc[(df["DataPedido"].dt.year == 2020) & (df["DataPedido"].dt.month == 1)]
vendas_janeiro_2020
```

```python
#Gráfico de vendas por região
df["VendasRegião"].value_counts(ascending=False).plot.bar(title="Vendas por Região");
```

```python
#Gráfico envolvendo produtos, quantidade, o custgo da unidade e o preço total de custo
df.groupby(df["Produto"])["Quantidade"].sum().plot()
plt.xlabel("CustoPorUnidade")
plt.ylabel("PreçoTotalDeCusto")
plt.legend();
```

```python
#Relação entre as entregas das vendas de janeiro e o valor da venda
plt.scatter(x=vendas_janeiro_2020["DataEntrega"], y=vendas_janeiro_2020["ValorTotalVenda"]);
```
